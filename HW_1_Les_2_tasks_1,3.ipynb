### 1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
1)	Найдите их сумму. (на листочке)
2)	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#Поскольку в первом предложении задания, относящемся к обеим задачам, сказано, 
#что даны два вектора, то во второй задаче реализуем написанную функцию для каждого из векторов.

def vector_magnitude(v):
    sum_sq = 0
    for i in range(len(v)):
        sum_sq += v[i]**2
    return math.sqrt(sum_sq)
```

```python
v1 = [20, 20, 0]
v2 = [0, 0, -20]

magn_v1 = vector_magnitude(v1)
magn_v2 = vector_magnitude(v2)

magn_v1, magn_v2
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

```python
# 1. Окружность
plt.gca().set_aspect('equal')
x = []
y1 = []
y2 = []
r = 5
for i in np.arange(-r, r, 0.0001):
    x.append(i)
    y1.append(math.sqrt(r**2 - i**2))
    y2.append(-math.sqrt(r**2 - i**2))
plt.plot(x,y1)   
plt.plot(x,y2)  
```

```python
#2. Эллипс

#параметрическое уравнение эллипса: x = u + a.cos(t); y = v + b.sin(t)

u=7      #x-координата центра
v=0.5    #y-координата центра
a=3      #радиус по x-оси
b=1.5    #радиус по y-оси

t = np.linspace(0, 5*math.pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid()
```

```python
#3. Гипербола

import decimal

xmin = -20
xmax = 20
dx = 0.1

xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)

ylist = 1 / xlist
plt.grid()
plt.plot(xlist, ylist)
plt.show()
```
