Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
k1, k2, k3 = 3, -4, 1/3
a1, a2, a3 = -3, 1/5, 2
b1, b2, b3 = 1/4, 3, 5
x = np.linspace(-2*np.pi, 3*np.pi, 201)
plt.plot(x, k1*np.cos(x - a1) + b1)
plt.plot(x, k2*np.cos(x - a2) + b2)
plt.plot(x, k3*np.cos(x - a3) + b3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

1) Напишите код, который будет переводить полярные координаты в декартовы. <p>
2) Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def decart2polar(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)*180/np.pi
    return (rho, phi)
```

```python
#phi = np.linspace(-3, 5, 1000)
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)

plt.polar(decart2polar(x,y)[1], decart2polar(x,y)[0])
```

### 4. 1)
Решите систему уравнений:<p>
y = x2 – 1 <p>
exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
def eq_sys(t):
    x, y = t
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) -1)

x = np.linspace(-3, 3, 250)
plt.plot(x, x**2 - 1)
plt.plot(x, 1 + (np.exp(x) - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

x1, y1 = fsolve(eq_sys, (-1,3))
print(f'Корень 1: {x1, y1}')

x2, y2 = fsolve(eq_sys, (2,2))
print(f'Корень 1: {x2, y2}')
```

### 4. 2)
Решите систему уравнений и неравенств: <p>
y = x2 – 1 <p>
exp(x) + x∙(1 – y) > 1

```python
#подставим y в неравенство и нарисуем график
x = np.linspace(-2, 4.5, 250)
plt.plot(x, np.exp(x) + x*(1-(x**2-1))-1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
#т.к. видно 4 места пересечения графиком оси x, следовтельно, есть 4 корня

def ineq_sys(x):
    return (np.exp(x) + x*(2 - x**2) - 1)

x1 = fsolve(ineq_sys, -2)
x2 = fsolve(ineq_sys, 2)
x3 = fsolve(ineq_sys, 0.1)
x4 = fsolve(ineq_sys, 6)
print(x1, x2, x3, x4)
```
